```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import norm
```

```python
Cutlet_data=pd.read_csv("/content/Cutlets.csv")
Cutlet_data
```

```python
unit_A=pd.Series(Cutlet_data.iloc[:,0])
unit_A
```

```python
unit_B=pd.Series(Cutlet_data.iloc[:,1])
unit_B
```

```python
p_value=stats.t.cdf(unit_A,unit_B)
p_value
```

```python
p_value=stats.ttest_ind(unit_A,unit_B)
p_value
```

```python
#our standard value of alpha=0.05....(5% standard deviation).
#so our p_value is 0.4722 so its greater than alpha value 0.05 so we fail to reject Null Hypothesis, we accept Null Hypothesis.
```
