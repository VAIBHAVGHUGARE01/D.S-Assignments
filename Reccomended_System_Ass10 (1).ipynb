```python
#Change structure of dataset so that we can compute the similarity score
user_books_df = Book.pivot_table(index= 'User_ID', columns= 'Book_Title', values= 'Book_Rating').fillna(0)
user_books_df
```

```python
#Impute those NaNs with 0 values- 0 for didnot watch movies
user_books_df.fillna(0,inplace=True)
user_books_df
```

**Calculate Cosine Similarity Between Users**

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
user_sim = 1 - pairwise_distances(user_books_df.values, metric= 'cosine')
user_sim
```

**Store the Result in DataFrame**

```python
user_sim_df=pd.DataFrame(user_sim)
user_sim_df #Here show Index Id
```

```python
#Now we transfer this index to column and Row then we understand which UserId for how much rating
user_sim_df.index=Book.User_ID.unique()
user_sim_df.columns=Book.User_ID.unique()
user_sim_df
```
